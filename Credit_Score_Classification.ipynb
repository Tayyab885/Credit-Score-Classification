### Exploratory Data Analysis

- #### Let's Visualize categorical features

```python
categorical_features = list(df.select_dtypes(include=['object']).columns)
categorical_features.remove('Name')
categorical_features.remove('Payment_Behaviour')
categorical_features.remove('Type_of_Loan')
categorical_features
```

```python
for col in categorical_features:
    c = df[col].value_counts()
    plt.figure(figsize=(14,6))
    sns.barplot(x=c.values,y=c.index)
    plt.show()
```

- #### Checking if occupation of the person affects credit scores

```python
import plotly.express as px
fig = px.box(x=df['Occupation'],color=df['Credit_Score'],title="Credit Scores Based on Occupation")
fig.show()
```

- #### Checking Credit Scores Based on Annual Income

```python
plt.figure(figsize=(14,6))
sns.boxplot(x="Credit_Score",y="Annual_Income",data=df)
plt.title("Credit Scores Based on Annual Income")
plt.xlabel("Credit Score")
plt.ylabel("Annual Income")
plt.show()
```

- According to the above visualization, the more you earn annually, the better your credit score is.

- #### Credit Scores Based on Monthly Inhand Salary

```python
plt.figure(figsize=(14,6))
sns.boxplot(x="Credit_Score",y="Monthly_Inhand_Salary",data=df)
plt.title("Credit Scores Based on Monthly Inhand Salary")
plt.xlabel("Credit Score")
plt.ylabel("Monthly Inhand Salary")
plt.show()
```

- Similarly like annual income, the more monthly in-hand salary you earn, the better your credit score will become.

- #### Credit Scores Based on Number of Bank Accounts

```python
sns.boxplot(x="Credit_Score",y="Num_Bank_Accounts",data=df)
```
